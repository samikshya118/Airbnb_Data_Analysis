## Missing Values Handling

```python
df['last review'] = pd.to_datetime(df['last review'],errors = 'coerce')

df.fillna({'reviews per month' : 0 , 'last review' : df['last review'].min()} , inplace = True)

df.dropna(subset = ['NAME','host name'],inplace = True)
```

```python
print(df.isnull().sum())
```

## Correct Data Types

```python
#ensuring all columns have the correct data types
# removing dollar signs in service fee and price and converting it to float
df['price'] = df['price'].replace('[\$,]','',regex = True).astype(float)
df['service fee'] = df['service fee'].replace('[\$,]','',regex = True).astype(float)
```

## Remove Duplicates

```python
df.drop_duplicates(inplace = True)
```

```python
print(df.info())
```

```python
df = df.drop(columns = ["license","house_rules"],errors = 'ignore')
```

## Room Type Analysis 

```python
# distribution of differrent rooms 
plt.figure(figsize = (8,5))
sns.countplot(x = 'room type',data = df, color = 'hotpink')
plt.title('Room Type Distribution')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()
```

Insight: The Airbnb listings are heavily dominated by Entire home/apt, followed by Private rooms, indicating a strong user preference for private accommodations. Shared rooms and hotel rooms are rare, showing minimal demand and confirming Airbnb’s focus on peer-to-peer residential stays rather than traditional hotels.

```python
plt.figure(figsize = (12,8))
sns.countplot(y='neighbourhood group',data=df,color='lightgreen',order=df['neighbourhood group'].value_counts().index)
plt.title('Number of listings by Neighbourhoos Group')
plt.xlabel('Count')
plt.ylabel('Neighborhood Group')
plt.show()
```

Airbnb listings are predominantly concentrated in Manhattan and Brooklyn, indicating these areas as the primary hubs of short-term rentals. Queens shows moderate activity, while Bronx and Staten Island have limited listings. The presence of inconsistent neighborhood labels highlights the need for data preprocessing before advanced analysis or modeling.

 ## Price vs Room Type

```python
plt.figure(figsize = (10,6))
sns.boxplot(x = 'room type' , y = 'price' , hue = 'room type' , data = df,palette = 'Set1')
plt.title('Price vs Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price ($)')
plt.legend(title = 'Room Type')
plt.show()
```

Room type strongly influences price

Entire home/apt has the highest variability (premium + budget listings)

Shared rooms tend to be the lowest-priced overall

Presence of wide ranges suggests outliers and dynamic pricing

## Reviews Over Time

```python
df['last review'] = pd.to_datetime(df['last review'])
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size()

plt.figure(figsize = (12,6))
reviews_over_time.plot(kind = 'line',color = 'red')
plt.title('Number of Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.show()
```

The line plot provides a clear visualization of how the number of reviews changes over time. It helps identify overall trends and noticeable spikes in review activity, which serve as indicators of fluctuations in booking demand and user engagement. Periods of high review counts may correspond to increased tourism or platform growth, while sudden drops could indicate external factors such as policy changes, market conditions, or incomplete data. This analysis is useful for understanding the temporal dynamics of Airbnb listing popularity and user behavior over time.
